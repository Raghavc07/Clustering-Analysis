# Task 1

Here for the given dataset we are using 2 clustering methods : Kmeans and Hierarchical .
We plot the 2D/3D scatter plot for the original class and also for the k means and hierarchical clustering .
To evaluate , we use external validation methods such as F1 and Accuracy score and analyse them. 





# DATA SET 1

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
df = pd.read_csv('Data1.csv')
df
```

```python
# Plotting based on original dataset

# DATA SET 1

import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection = '3d')

# Data for three-dimensional scattered points
xdata = df['X1']
ydata = df['X2']
zdata = df['X3']
hue = df['Class']

ax.scatter3D(xdata, ydata, zdata, c=hue);

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

# ax.view_init(elev=0, azim=60) # to rotate the plot
```

K-Means

```python
# Getting clusters from K-Means Algo

# Within cluster evaluation
from sklearn.cluster import KMeans

# creating empty model of K-means
model_k = KMeans(n_clusters=7)

# predict the labels of clusters
model_k.fit(df)

k_predict = model_k.predict(df)
k_predict  # labels
```

```python
# adding predicted value to our original dataframe
df['K_Means Prediction'] = k_predict
print(df)

# adding 1 to the Predicted class to match with the  original dataset
df['K_Means Prediction'] = df['K_Means Prediction'] + 1
print(df)
```

```python
# plotting based on K-Means Clusters

fig = plt.figure()
ax = plt.axes(projection = '3d')

# Data for three-dimensional scattered points
xdata = df['X1']
ydata = df['X2']
zdata = df['X3']
hue = df['K_Means Prediction']
ax.scatter3D(xdata, ydata, zdata, c=hue);

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

# ax.view_init(elev=0, azim=60) # to rotate the plot
```

```python
from sklearn import metrics

# External Validation - Accuracy
print('accuracy: ', metrics.accuracy_score(df["Class"], df["K_Means Prediction"]))
```

```python
# External Validation - F1-score
print('f1 score: ', metrics.f1_score(df["Class"], df["K_Means Prediction"], average='micro'))
```

Hierarchical

```python
from sklearn.cluster import AgglomerativeClustering
modelh = AgglomerativeClustering()
```

```python
# Dropping the K_Means Prediction to get back original dataset
df = df.drop("K_Means Prediction", axis=1)
```

```python
df # original dataset
```

```python
clusters = 7
hc = AgglomerativeClustering(n_clusters=clusters, affinity = 'euclidean', 
                             linkage = 'ward')

# save clusters for chart
h_predict = hc.fit_predict(df)
h_predict
```

```python
# adding a new column to the original dataset
df['H_Clusters'] = h_predict

df['H_Clusters'] = df['H_Clusters'] + 1
df
```

```python
# plotting based on Hierarchical Clusters

fig = plt.figure()
ax = plt.axes(projection = '3d')

# Data for three-dimensional scattered points
xdata = df['X1']
ydata = df['X2']
zdata = df['X3']
hue = df['H_Clusters']
ax.scatter3D(xdata, ydata, zdata, c=hue);

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

# ax.view_init(elev=0, azim=60) # to rotate the plot
```

```python
#Removing Special Characters
df['Business Tax Rate'] = df['Business Tax Rate'].apply(lambda x: x.strip("%"))
df['GDP'] = df['GDP'].apply(lambda x: x.strip("$").replace(",", ""))
df['Health Exp/Capita'] = df['Health Exp/Capita'].apply(lambda x: x.strip("$").replace(",", ""))
```

```python
#Scaling the data since we have huge range of values acorss the columnns
scaler = StandardScaler()
features_std = scaler.fit_transform(features)
```

Hierarchical Clustering

```python
#Plotting Dendrogram
plt.figure(figsize=(30,15))

dendrogram = sch.dendrogram(sch.linkage(features_std, method='ward'), labels=labels)
# plt.axhline(y=11, color='r', linestyle='--')
```

```python
#Elbow Method
from yellowbrick.cluster import KElbowVisualizer
model = AgglomerativeClustering()

visualizer = KElbowVisualizer(model, k=(1,11), timings=False)
visualizer.fit(features_std)
visualizer.show()
```

Elbow Method giving number of clusters as 3

```python
#Sillouette Coefficients 
silhouette_coefficients_hc = []
for n_clusters in range(2, 11):
    clusterer = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='ward')
    cluster_labels = clusterer.fit_predict(features_std)
    score = silhouette_score(features_std, cluster_labels)
    silhouette_coefficients_hc.append(score)

plt.plot(range(2, 11), silhouette_coefficients_hc)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

Sillouette Coefficient giving highest value at 2 but high value for 3 as well

```python
#Selecting Clusters as 3
clusters = 3
hc = AgglomerativeClustering(n_clusters=clusters, affinity = 'euclidean', linkage = 'ward')

# save clusters for chart
y_hc = hc.fit_predict(features_std)
```
